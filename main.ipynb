```python
#ols 돌리기

xdata = ''


df_train = pd.DataFrame(index=range(0,len(group_unique_path[34])))
df_train = train_numeric[train_numeric['Id'].isin(group_unique_path[34])]




df_train = df_train.dropna(axis=1)



df_test = df_train[df_train['Id'].isin(group_unique_path[34][491:541])]
# df_diff = df_test
# df_diff.to_csv("C:\\Users\\wg999\\OneDrive\\바탕 화면\\업무\\프로젝트\\dataset\\df_diff.csv", header= True, index=False)
del df_test['output']

idx_temp = df_train[df_train['Id'].isin(group_unique_path[34][491:541])].index
df_train = df_train.drop(idx_temp)





for col in df_train.columns[1:-1]:
    if col != df_train.columns[-2]:
        xdata = xdata + col + ' + '
    else:
        xdata = xdata + col
        
import statsmodels.api as sm

reg = sm.OLS.from_formula("output ~ " + xdata, df_train).fit()
reg.summary()

```
